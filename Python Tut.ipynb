```
def Factorial(NumFact=1):
    product=float(1)
    a=1
    for i in range(1,NumFact+1):
        product=product*a
        a+=1
    return product
```

```
print(Factorial(10))
```

```
def Factorial_Recursive(NUM=1,a=1):
    
    if NUM>1:
        a=a*NUM
        return Factorial_Recursive(NUM-1,a)
    else:
         return a   
```

```
print(Factorial_Recursive(8))
```

```
def expo(x=1,n=1):
    x=float(x)
    n=int(n)
    final=x**n/Factorial(n)
    return final
```

```
print(expo(5,2))
```

```
def exponential(x=1,n=100):
    sum=float(1)
    for i in range(n):
        sum=sum+float(expo(x,n))
        n-=1
    return sum
```

```
print(exponential(2))
```

```
v1=exponential(1.3)*exponential(-1.04)
v2=exponential(0.26)
print(round(v1-v2,10))
```

Evidently, it is an Exponential Function

```
def expo_epsilon(x=1,epsilon=0.001):
    sum=float(1)
    i=1
    while True:
        sum=sum+float(expo(x,i))
        i+=1
        if expo(x,i)<epsilon:
            break
    return sum
```

```
def matrix_multiply(A,n):
    import copy
    p=copy.deepcopy(A)
    r=[
       [0,0],
       [0,0]
    ]
    if n<=1:
        return p        
    for i in range(n-1):
        r[0][0]= p[0][0]*A[0][0] + p[0][1]*A[1][0]
        r[0][1]= p[0][0]*A[0][1] + p[0][1]*A[1][1]
        r[1][0]= p[1][0]*A[0][0] + p[1][1]*A[1][0]
        r[1][1]= p[1][0]*A[0][1] + p[1][1]*A[1][1]
        p=copy.deepcopy(r)
    return r

    
```

```
with open("poem.txt","r") as file_open:
    line_list=file_open.readlines()
b=""
char_replace=['\n',',','"','-',':','(',')','.']
for line in line_list:
    line=line.lower()
    for s in char_replace:
        line = line.replace(s," ")
    fixed_line = line.strip()
    b+=" " + fixed_line
word_list = b.split()
word_dict = {}
for word in word_list:
    if word in word_dict:
        word_dict[word]+=1
    else:
        word_dict[word]=1
print(word_dict)
    
    
```

```
top_ten=sorted(word_dict.items(),key=lambda item: item[1],reverse=True)
print(top_ten)
top_ten[:12]
```

```
with open("Top Ten Words.txt","w") as top_file:
    for item in top_ten[:10]:
        top_file.write(item[0]+"\n")
```

```
two_gram_dict={}
for i in range(len(word_list)-1):
    two_gram = word_list[i] + " " + word_list[i+1]
    if two_gram in two_gram_dict:
        two_gram_dict[two_gram]+=1
    else:
        two_gram_dict[two_gram]=1
two_gram_dict
```
