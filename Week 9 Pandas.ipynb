```
diameter = pd.Series([4879.4, 12104, 12742, 6779, 139820, 116460, 50724, 49244, 2376.6], 
                 index=['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune', 'Pluto'])
```

```
density_kgm3=(10**15)*6*mass/(math.pi*(diameter.pow(3)))
```

```
density_kgm3
```

```
mass["PlanetX"]=6.23
diameter["PlanetY"]=12341
```

```
density_kgm3[pd.isnull(density_kgm3)]=np.mean(density_kgm3)
```

```
%%time
for key in density_kgm3.index:
    if key in mass.index:
        print(key, "has mass of", mass[key])
    else:
        print(key,"is missing")
```

```
density_dict={}
for key in density_kgm3.index:
    density_dict[key]=density_kgm3[key]
mass_dict={}
for key in mass.index:
    mass_dict[key]=mass[key]
```

```
%%time
for key in density_dict:
    if key in mass_dict:
        print(key, "has mass of", mass_dict[key])
    else:
        print(key,"is missing")
```

```
%%time
np.std(density_kgm3)
```

```
%%time
np.std(density_dict.values())
```
