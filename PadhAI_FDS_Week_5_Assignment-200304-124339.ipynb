````
# function to return the slope of a given function
def Slope_finder(tuple_contain_function):
    x_org=float(tuple_contain_function[1])
    func=tuple_contain_function[0]
    func_org=func(x_org)
    change=0.000001
    x_new=x_org+change
    func_new=func(x_new)
    slope=(func_new - func_org)/change
    answer=(round(slope,3),x_org,func_org,change)
    return answer
````

````
Slope_finder((linear_fn,4))
````

````
def find_derivative(data):
    func, x_a = data
    delta_x = 0.000001
    f_x_a = func(x_a)
    f_delta_x = func(x_a + delta_x)
    return ((f_delta_x - f_x_a)/delta_x, x_a, f_x_a, delta_x)
````

````
find_derivative((linear_fn,4))
````

````
def factorial(n):
    product=1
    for i in range(n):
        product=product*(i+1)
    return product
````

````
def exponential(n,a):
    sum=float(0)
    for i in range(n):
        sum+=a**i/factorial(i)
    return sum
````

````
# tuple of exponential functions
def tuple_of_expo(n,a):
    temp_list=[]
    for i in range(n):
        temp_list.append(exponential(i+1,a))
    return tuple(temp_list)
````

````
def marks_operation(dictionary_of_one_student):
    marks_list=dictionary_of_one_student["Marks"]
    alpha_list=[0]
    beta=0.99
    for i in range(10):
        alpha_list.append(beta*alpha_list[i]+(1-beta)*marks_list[i])
    return alpha_list[1:]
````

````
def is_prime(num):
    counter=0
    if num==1:
        return False
    for i in range(2,num-1):
        if num%i==0:
            counter+=1
    return True if counter==0 else False
````

````
fibonacci=[1,1]
while fibonacci[-1]+fibonacci[-2]<=100:
    fibonacci.append(fibonacci[-1]+fibonacci[-2])
fibonacci
````

````
dict_category={}
for i in range(1,101):
    if i not in dict_category:
        dict_category[i]=[]
    if i**2<=100:
        dict_category[i**2]=["Square"]
    if i in fibonacci:
        dict_category[i].append("Fibonacci")
    if is_prime(i):
        dict_category[i].append("Prime")
dict_category
````

````
def seq_list(Dictionary,Category):
    numbers_list=[]
    for i in range(1,len(Dictionary)+1):
        if Category in Dictionary[i]:
            numbers_list.append(i)
    return tuple(numbers_list)
````

````
primes=seq_list(dict_category,"Square")
print(primes)
````
