## Mean Row

```
def Create_Mean(Dataframe):
    nDataframe = Dataframe.copy()
    nDataframe.loc['Mean'] = [np.mean(nDataframe.iloc[:,i]) for i in range(len(nDataframe.columns))]
    return nDataframe
```

## Shortener

```
planets_full = sns.load_dataset('planets')
```

```
def shortener(s):
    return ''.join([x[0] for x in s.split()])
```

```
planets_full['Short Method'] = planets_full['method'].apply(shortener)
```

```
planets_full
```

## GroupBy

```
pd.DataFrame({"Count":planets_full.groupby("method")["method"].count(),"Mean Distance":planets_full.groupby("method")['distance'].mean()})
```

# Tasks

Tasks on the NIFTY datasets:

1. In 2019, in how many days was the NIFTY50 volatile (high > 105% of low)

2. In 2019, in how many days was the NIFTYNEXT50 volatile (high > 105% of low)

3. In 2019, how many days belonged to the four classes NIFTY50 volatile / non-volatile and NIFTYNext50 volatile / non-volatile

4. Compute the mean, median, std, var of closing values for each weekday in NIFTY50 for 2019

5. Compute the mean, median, std, var of closing values for each month in NIFTY50 for 2019

6. On the days in which NIFTY50 closed higher than the open, what was the mean of (close - open) for NIFTYNext50

7. In 2019, how many days had the day's high lower than the previous day's low in NIFTY50

8. In 2019, on how many days did the day's close exceed the 30 day moving average in NIFTY50 (exclude first month)

```
nifty502018 = pd.read_csv('NIFTY50_2018.csv',index_col=0,infer_datetime_format=True,parse_dates=True)
nifty502019 = pd.read_csv('NIFTY50_2019.csv',index_col=0,infer_datetime_format=True,parse_dates=True)
niftynext502019 = pd.read_csv('NIFTYNext50_2019.csv',index_col=0,infer_datetime_format=True,parse_dates=True)
```

```
nifty502019['Volatile'] = (nifty502019["High"] > (1.05 * nifty502019["Low"]))
```

```
niftynext502019['Volatile'] = (niftynext502019["High"] > (1.05 * niftynext502019["Low"]))
```

```
print("Nifty 50 Volatile:",np.sum(nifty502019['Volatile']),'\n',
      "Nifty 50 Non-Volatile:",np.sum(~nifty502019['Volatile']),'\n',
      "Nifty Next 50 Volatile:",np.sum(niftynext502019['Volatile']),'\n',
      "Nifty Next 50 Non-Volatile:",np.sum(~niftynext502019['Volatile']),'\n'
      )
```

```
day_of_week = pd.Series(map(lambda x: x.dayofweek,nifty502019.index),index=nifty502019.index).rename("Day of Week")
month = pd.Series(map(lambda x: x.month,nifty502019.index),index=nifty502019.index).rename("Month")
```

```
dow = pd.Series(nifty502019.index.dayofweek,index = nifty502019.index).rename('Day of Week')
monthname = pd.Series(nifty502019.index.month,index = nifty502019.index).rename('Month')
```

```
dw = pd.concat([nifty502019,day_of_week],axis=1)
mo = pd.concat([nifty502019,month],axis=1)
```

```
pd.DataFrame({"Mean" : dw.groupby("Day of Week")["Close"].mean(),
      "Median" : dw.groupby("Day of Week")["Close"].median(),
      "Std" : dw.groupby("Day of Week")["Close"].std(),
      "Var" : dw.groupby("Day of Week")["Close"].var(),
})
```
