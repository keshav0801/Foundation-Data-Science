# Exercise

```
random_pts=np.random.rand(1000000,2)
```

```
%%time
inside=0
outside=0
for row in random_pts:
    if row[0]**2 + row[1]**2 < 1:
        inside+=1
    else:
        outside+=1
area_outside=outside/(inside+outside)
print(area_outside)
```

```
samplepts=10000000
dimensions=2
randompts=np.random.rand(samplepts,dimensions)
```

```
%%time
outside=0
dfo=np.zeros(samplepts)
for i in range(samplepts):
    for j in range(dimensions):
        dfo[i]+=randompts[i,j]**2
    if dfo[i]>1:
        outside+=1
print(outside/samplepts)
```

```
def area_outside(sample_pts,dimensions):
    randompts=np.random.rand(sample_pts,dimensions)
    return np.sum(np.sqrt(np.sum(randompts*randompts,axis=1)) >1)/sample_pts
```

```
%%time
for i in range(2,11):
    print(i," :",area_outside(10000000,i))
```

```
%%time
print(6," :",area_outside(10000000,6))
```
